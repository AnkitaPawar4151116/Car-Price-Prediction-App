```python
import numpy as np
import pandas as pd
```

```python
ds = pd.read_csv('data.csv')
print(ds.to_string())
```

```python
ds.info()
```

```python
ds.describe()
```

```python
ds['year'].unique()
```

```python
#dropping null cells
ds.dropna(inplace=True)
```

```python
#convert data type of year column object to int
ds['year']= pd.to_numeric(ds['year'], errors='coerce')
```

```python
ds.info()
```

```python
print(ds.to_string())
```

```python
# for price
ds['Price'].unique()
```

```python
ds['Price']= ds['Price'].str.replace(',','')
```

```python
ds['Price']= pd.to_numeric(ds['Price'], errors='coerce')
```

```python
ds['Price'].unique()
```

```python

ds['kms_driven'] = ds['kms_driven'].str.replace(',','').str.replace(' kms','')
```

```python
ds['kms_driven']= pd.to_numeric(ds['kms_driven'], errors='coerce')
```

```python
#to remove outliers to kns_driven
ds= ds[ds['kms_driven'] >= 10000]
ds= ds[ds['kms_driven'] <= 200000]
```

```python
ds.info()
```

```python
sorted(ds['kms_driven'].unique())
```

```python
ds['fuel_type'].unique()
```

```python
ds.info()
```

```python
ds['company'].unique()
```

```python
#remove company names from name column
for i in ds.index:
    company= ds.loc[i,'company']
    name= ds.loc[i,'name']
    name= name.replace(company,'').strip()
    ds.loc[i,'name']= name
```

```python
print(ds.to_string())
```

```python
ds.info()
```

```python
sorted(ds['year'].unique())
```

```python
ds.duplicated().sum()
```

```python
ds.drop_duplicates(inplace=True)
```

```python
ds.info()
```

```python
ds['Price'].isnull().sum()
```

```python
ds= ds.dropna(subset= ['Price'])
```

```python
print(ds.to_string())
```
